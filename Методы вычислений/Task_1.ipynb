Вариант 5

f(x)= cos(x), узлы: -0.6, -0,5, -0,3, -0,2,-0.1, 0. Точка интерполирования -0.4. Точное значение функции f(-0.4) = 0.8

```python
import pandas as pd
import math as ms
import numpy as np
from sympy import symbols, solveset, Interval, Min, Max
import sympy as sym
```

```python
data1 = {'x':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],'f(x)':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],'р.р 1-ого п.':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],'р.р. 2-ого п.':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],'р.р 3-его п.':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]}
table1 = pd.DataFrame(data1)
table1.loc[0,'x']=-0.5
table1.loc[2,'x']=-0.3
table1.loc[4,'x']=-0.6
table1.loc[6,'x']=-0.2

for i in range(7):
    if i% 2 == 0:
         table1.loc[i,'f(x)']= round(ms.cos(table1['x'][i]),9)
for i in range(3):
    table1.loc[2*i+1,'р.р 1-ого п.'] = round((table1['f(x)'][2*i+2]-table1['f(x)'][2*i])/(table1['x'][2*i+2]-table1['x'][2*i]),4)
for i in range(1,3):
    table1.loc[2*i,'р.р. 2-ого п.'] = round((table1['р.р 1-ого п.'][2*i+1]-table1['р.р 1-ого п.'][2*i-1])/(table1['x'][2*i+2]-table1['x'][2*i-2]),4)
table1.loc[3,'р.р 3-его п.']= round((table1['р.р. 2-ого п.'][4]-table1['р.р. 2-ого п.'][2])/(table1['x'][6]-table1['x'][0]),4) 
display(table1)
```

```python
data2={'Узлы в порядке очередности их использования':[0,0,0,0],'Pi(-0.4)-значение многочлена в точке интерполирования':[0,0,0,0],'f(-0,4)-Pi(-0.4)-фактическая погрешность':[0,0,0,0],'Mi+1 -оценка модуля произв.':[0,0,0,0],'Ri(-0.4)-оценка погрешности':[0,0,0,0]}
table2 = pd.DataFrame(data2)
for i in range(4):
    table2.loc[i,'Узлы в порядке очередности их использования'] = table1['x'][2*i]
table2.loc[0,'Pi(-0.4)-значение многочлена в точке интерполирования'] = table1['f(x)'][0]
table2.loc[1,'Pi(-0.4)-значение многочлена в точке интерполирования'] = table2['Pi(-0.4)-значение многочлена в точке интерполирования'][0] + table1['р.р 1-ого п.'][1]*(-0.4-table1['x'][0])
table2.loc[2,'Pi(-0.4)-значение многочлена в точке интерполирования'] = table2['Pi(-0.4)-значение многочлена в точке интерполирования'][1] + table1['р.р. 2-ого п.'][2]*(-0.4-table1['x'][0])*(-0.4-table1['x'][2])
table2.loc[3,'Pi(-0.4)-значение многочлена в точке интерполирования'] = table2['Pi(-0.4)-значение многочлена в точке интерполирования'][2] + table1['р.р 3-его п.'][3]*(-0.4-table1['x'][0])*(-0.4-table1['x'][2])*(-0.4-table1['x'][4])
for i in range(4):
    table2.loc[i,'f(-0,4)-Pi(-0.4)-фактическая погрешность'] = 0.8 - table2['Pi(-0.4)-значение многочлена в точке интерполирования'][i]
a = min(-0.5,-0.3,-0.6,-0.2,-0.4)
b = max(-0.5,-0.3,-0.6,-0.2,-0.4)
x = symbols('x')
f = sym.cos(x)
for i in range(4):
    f = sym.diff(f,x)
    table2.loc[i,'Mi+1 -оценка модуля произв.'] = max(abs((f).subs(x,a)),abs((f).subs(x,b)))
f = 1
for i in range(4):
    f = f * (-0.4-table1['x'][2*i])/(i+1)
    table2.loc[i,'Ri(-0.4)-оценка погрешности'] = table2['Mi+1 -оценка модуля произв.'][i]*f
display(table2)
```

```python
P3 = table1['f(x)'][0]
P3 += table1['р.р 1-ого п.'][1]*(x+0.5)
P3 += table1['р.р. 2-ого п.'][2]*(x+0.5)*(x+0.3)
P3 += table1['р.р 3-его п.'][3]*(x+0.5)*(x+0.3)*(x+0.6)
Fact_err = 0.8 - P3.subs(x, -0.4)
print("Полином P3: ", P3)
print("Погрешность", Fact_err)

Solutions = sym.solve(P3 - 0.8, x)
for x1 in Solutions:
    if type(x1) != sym.core.add.Add:
        print(x1)
```
